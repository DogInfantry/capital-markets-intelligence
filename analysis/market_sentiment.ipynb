```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime

sns.set_theme(style='whitegrid', palette='deep')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 11
```

## 1. Load Data

```python
# Load IPO data
ipo_df = pd.read_csv('../data/raw/ipo_data_raw.csv', parse_dates=['ipo_date'])
print(f'IPOs loaded: {len(ipo_df)}')
ipo_df.head()
```

```python
# Load M&A data
mna_df = pd.read_csv('../data/raw/mna_data_raw.csv', parse_dates=['announced_date'])
print(f'M&A deals loaded: {len(mna_df)}')
mna_df.head()
```

## 2. IPO Market Trends

```python
# IPO first-day returns distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].bar(range(len(ipo_df)), ipo_df['first_day_return_pct'].sort_values(ascending=False),
            color=['#2ecc71' if x > 0 else '#e74c3c' for x in ipo_df['first_day_return_pct'].sort_values(ascending=False)])
axes[0].set_title('IPO First-Day Returns (%)')
axes[0].set_ylabel('Return %')
axes[0].axhline(y=0, color='black', linewidth=0.5)

axes[1].bar(range(len(ipo_df)), ipo_df['total_return_pct'].sort_values(ascending=False),
            color=['#2ecc71' if x > 0 else '#e74c3c' for x in ipo_df['total_return_pct'].sort_values(ascending=False)])
axes[1].set_title('IPO Total Returns to Date (%)')
axes[1].set_ylabel('Return %')
axes[1].axhline(y=0, color='black', linewidth=0.5)

plt.tight_layout()
plt.savefig('../output/ipo_returns.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# IPO deal size by sector
sector_summary = ipo_df.groupby('sector').agg(
    count=('company', 'count'),
    total_deal_size=('deal_size_usd', 'sum'),
    avg_first_day_return=('first_day_return_pct', 'mean')
).sort_values('total_deal_size', ascending=False)

print('IPO Activity by Sector:')
sector_summary
```

## 3. M&A Deal Flow Analysis

```python
# M&A deal value by sector
fig, ax = plt.subplots(figsize=(12, 6))
sector_deals = mna_df.groupby('sector')['deal_value_usd_bn'].sum().sort_values(ascending=True)
sector_deals.plot(kind='barh', ax=ax, color='#3498db')
ax.set_title('M&A Deal Value by Sector ($B)')
ax.set_xlabel('Deal Value ($B)')
plt.tight_layout()
plt.savefig('../output/mna_by_sector.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Deal status breakdown
status_counts = mna_df['status'].value_counts()
fig, ax = plt.subplots(figsize=(8, 5))
colors = {'Completed': '#2ecc71', 'Pending Regulatory': '#f39c12', 'Blocked (DOJ)': '#e74c3c', 'Blocked (CFIUS)': '#c0392b'}
status_counts.plot(kind='bar', ax=ax, color=[colors.get(s, '#95a5a6') for s in status_counts.index])
ax.set_title('M&A Deal Status')
ax.set_ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- International IPOs (India, UAE) showing increasing scale
