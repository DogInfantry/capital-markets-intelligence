```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Load case studies
cs_df = pd.read_csv('../data/processed/case_studies.csv')
print(f'Case studies loaded: {len(cs_df)}')
cs_df.head()
```

```python
# Deal size distribution by tier
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

tier_counts = cs_df['size_tier'].value_counts()
tier_counts.plot(kind='pie', ax=axes[0], autopct='%1.0f%%', startangle=90)
axes[0].set_title('Deals by Size Tier')
axes[0].set_ylabel('')

cs_df.sort_values('deal_value_usd_bn', ascending=True).plot(
    kind='barh', x='deal_name', y='deal_value_usd_bn', ax=axes[1],
    color='#3498db', legend=False
)
axes[1].set_title('Deal Value ($B)')
axes[1].set_xlabel('$B')

plt.tight_layout()
plt.savefig('../output/deal_overview.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Research relevance ranking
fig, ax = plt.subplots(figsize=(10, 8))
ranked = cs_df.sort_values('research_relevance_score', ascending=True)
colors = ['#e74c3c' if s >= 8 else '#f39c12' if s >= 6 else '#2ecc71' for s in ranked['research_relevance_score']]
ax.barh(ranked['deal_name'], ranked['research_relevance_score'], color=colors)
ax.set_title('Research Relevance Score (GS/JPM/DB)')
ax.set_xlabel('Score (1-10)')
ax.set_xlim(0, 10)
plt.tight_layout()
plt.savefig('../output/research_relevance.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Key Observations

- Mega-deals ($50B+) concentrated in Energy and Technology
- Regulatory risk is the dominant theme across blocked/pending deals
- Cross-border deals face additional CFIUS/antitrust scrutiny
- PE-backed take-privates active in fintech and healthcare
